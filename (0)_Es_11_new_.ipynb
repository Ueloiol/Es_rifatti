```python

class Utente:
    def __init__(self, Username, Password):
        self.Username = Username
        self.Password = Password

class Ferie(Utente):
    def __init__(self, Username, Password, spesa, tempo, vista):     
        super().__init__(Username, Password)
        self.spesa = spesa
        self.tempo = tempo
        self.vista = vista

class Gestionale_Ferie:
    Elenco = []

    # menu più generale, posso scegliere che operazione effettuare
    def switch(self):
        while True:
          print('\nCosa vuoi fare?')
          print('(1) Esci')
          print('(2) (Solo Admin) visualizza tutti gli utenti')
          print('(3) Registrati')
          print('(4) Login')
          scelta = int(input(''))

          if scelta == 1:
            break

          elif scelta ==2:
            self.admin_visualizzazione() 

          elif scelta ==3:
            self.registra_utente()       

          elif scelta == 4:
            self.login_utente()


    def registra_utente(self):
        nome = input('\ninserisci lo username: ')
        password = input('inserisci la password: ')

        # check per capire se il nome utente è disponibile
        counter = 0
        for i in range(len(self.Elenco)):
          if nome == self.Elenco[i].Username :
              counter += 1
        
        # se il nome utente è disponibile, aggiungo le informazioni all'elenco
        if counter== 0:
          print('\nIl nome utente selezionato è disponibile, inserisci le altre informazioni:')
          spesa = int(input('spesa per la vacanza: '))
          tempo = input('inserisci il tempo di permanenza (es. 7giorni):')
          vista = bool(int(input('vista mare (0= False, 1=True)): ')))
          x = Utente(nome, password)
          y = Ferie(nome, password, spesa, tempo, vista)
          self.Elenco.append(y)
        
           # se l'utente è già registrato, stampo un messaggio di errore
        else:
          print('\nUtente già presente nel gestionale, effettua il login')
    
    # visualizzazione di tutti gli utenti
    # ovviamente in un'applicazione reale questa opzione non è possibile darla
    # ai singoli utenti.
    def admin_visualizzazione(self):

        # se l' elenco è vuoto, mi viene scritto un messaggio d'errore
        if len(self.Elenco)==0:
          print('\nNon sono presenti utenti registrati')

        # se l'elenco non è vuoto vengono stampate le informazioni
        elif len(self.Elenco)>0:
          for i in range(len(self.Elenco)):
            print('\n  Username: ', self.Elenco[i].Username)
            print('  Password: ', self.Elenco[i].Password)
            print('     Spesa: ', self.Elenco[i].spesa)
            print('Permanenza: ', self.Elenco[i].tempo)
            print('Vista mare: ', self.Elenco[i].vista)

    def login_utente(self):
        nome = input('\ninserisci lo username: ')
        password = input('inserisci la password: ')

        # se c'è un utente nell'elenco con stesso username e password allora
        # flag+=1 (sarà 0 o 1 dato che non sono ammessi duplicati)
        # index = i (mi dice qual è la posizione dell'utente nell'elenco)
        flag=0
        index = 0
        for i in range(len(self.Elenco)):
          if nome == self.Elenco[i].Username and password == self.Elenco[i].Password:
            flag+=1
            index = i
        
        # se c'è corrispondenza viene stampato un messaggio
        # ed entro nel menu post login tramite switch_post_login 
        if flag==1:
          print('\nlogin effettuato con successo!')
          self.switch_post_login(index)               # qui entro nel nuovo menu
        
        # se rimane flag=0 significa che non c'è corrispondenza
        else:
          print('\nspiacente, credenziali errate')

    def switch_post_login(self, index):
        while True:
          print('\nseleziona cosa vuoi fare: ')
          print('(1) visualizza le informazioni utente')
          print('(2) modifica le informazioni utente')
          print('(3) logout')
          scelta = int(input(''))

          if scelta==1:
            print('\n  Username: ', self.Elenco[index].Username)
            print('  Password: ', self.Elenco[index].Password)
            print('     Spesa: ', self.Elenco[index].spesa)
            print('Permanenza: ', self.Elenco[index].tempo)
            print('Vista mare: ', self.Elenco[index].vista)

          # se scelgo di modificare le informazioni dell'utente, entro nel nuovo 
          # metodo self.modifica che mi permette di modificarle.
          elif scelta ==2:
            self.modifica(index)

          elif scelta == 3:
            break

    def modifica(self, index):
        while True:
          print('\nseleziona quale voce vuoi modificare: ')
          print('(1) spesa')
          print('(2) tempo di viaggio')
          print('(3) vista mare')
          print('(4) torna indietro')
          scelta = int(input(''))

          if scelta==1:
            spesa_new = int(input('\nseleziona il nuovo valore per la spesa: '))
            self.Elenco[index].spesa = spesa_new 

          elif scelta ==2:
            tempo_new = input('seleziona il nuovo valore per permanenza: ')
            self.Elenco[index].tempo = tempo_new 

          elif scelta == 3:
            vista_new = bool(int(input('seleziona il nuovo valore per vista mare (0=False, 1=True): ')))
            self.Elenco[index].vista = vista_new 

          elif scelta == 4:
            break

```
